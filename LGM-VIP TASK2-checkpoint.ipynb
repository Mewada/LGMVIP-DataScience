```python
#importing the libraries:
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# Reading the image and resize it give to color then show
image = cv2.imread('Army.jpg')
image = cv2.resize(image, (600, 800))
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.title('Image')
plt.show()
```

```python
#we will create a new image by converting the original image to greyscale:
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("Army", gray_image)
cv2.waitKey()
```

```python
#the next step is to invert the new grayscale image:
inverted_image = 255 - gray_image
cv2.imshow("Inverted", inverted_image)
cv2.waitKey()
```

```python
#the process is to blur the image by using the Gaussian Function in OpenCV:
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

```python
#the final step is to invert the blurred image, then we can easily convert the image into a pencil sketch:
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2.imshow("Sketch", pencil_sketch)
cv2.waitKey(0)
```

```python
#Have a look at both the original image and the pencil sketch 
cv2.imshow("original image", image)
cv2.imshow("pencil sketch", pencil_sketch)
cv2.waitKey(0)
```

```python
#Final output
plt.figure(figsize=(8,8))
plt.imshow(pencil_sketch, cmap="gray")
plt.title('Sketching of Image', fontsize=26)
plt.gca().axes.get_yaxis().set_visible(False)
plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
```
